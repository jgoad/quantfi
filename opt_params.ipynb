```python
def kscomp(mu_sigma,trc_func,t_series,p_series,bar_num,*faux_args):
    
    mu,sigma = mu_sigma
    trc = trc_func('price',t_series,p_series,mu,sigma,*faux_args)
    fds_trc = np.array(ufuncs.get_fds(trc))
    cnt_trc,mkr_trc = np.histogram(fds_trc,bar_num)
    norm_hist = np.histogram(fds_trc,bar_num)[0]
    norm_cnt = normpdf(mkr_trc,0,np.std(fds_trc))
    norm_cnt = norm_cnt*max(cnt_trc)/max(norm_cnt)
    return ks_2samp(norm_cnt, cnt_trc)[0]

def optmz_params(trc_func,t_series,p_series,prms_init,hist_gran,*faux_args):
    '''Optomize the parameters of the trace function
    
    trc_func = Use etrace or atrace to act on dat_series and t_series
    t_series = The time series corrisponding to dat_series
    p_series = The data set which will be traced
    prms_init = Initial guess for the parameters mu and sigma
    hist_gran = Granularity of the trace's histogramed fds
    '''
    return minimize(kscomp,prms_init,(trc_func,t_series,p_series,100),method='CG')
```
