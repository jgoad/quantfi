```python
import random
import textwrap
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from math import log
from math import exp
```

```python
def wiener(n, dt, t_init=0, w_init=0.0):
    """Returns one realization of a Wiener process with n steps of length dt.
    The time and Wiener series can be initialized using t_init and w_init respectively."""
    n+=1
    t_series = np.arange(t_init,n*dt,dt)
    h = t_series[1]-t_series[0]
    z = np.random.normal(0.0,1.0,n)
    dw = np.sqrt(h)*z
    dw[0] = w_init
    w_series = dw.cumsum()
    return t_series, w_series
```

#Exponential Brownian Motion With Drift

```python
def euler_step(n, dt, a, b, m=1, mu=0, sigma=1, w_init=0, p_init=1, chain_length=0): 
    '''Moves a price series forward through n time steps using Euler's method.
    
    Details
    -------
    Price trajectories have constant drift terms and growth is reduced with noise.
    Returns a price series along with its corisponding wiener and time series.
    
    Argumets
    --------
    n = the number of steps taken
    dt = the size of the time steps
    a = constant controling the significance of price drift
    b = constant controling the significance of the noise
    m = the default case refers to standard geometric brownian motion
    mu = mean of the normal distribution
    sigma = standard deviation of the normal distribution
    w_init = Initial value of the wiener series
    p_init = Initial value of the price series. If chaining, give p_init of the
        previously existing series which the returned one will be linked to.
    chain_length = The number of time steps in the previously existing series which
        the returned ones will be linked to. (default assumes no chaining)
    '''
    w_t = w_init
    p_t = p_init
    if chain_length>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [w_init]
        p_series = [p_init]
    for i in range(n):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        # Incrament p(t) with dp as given by the DE.
        # Incrament w(t) with dw as given by the normal distribution
        p_t += (a*dt+b*nu)*p_t**(m)
        w_t += nu
        # Append new values to each series;
        w_series.append(w_t)
        p_series.append(p_t)
    return t_series,w_series,p_series

def analy_step(n, dt, a, b, m=1, mu=0, sigma=1, w_init=0, p_init=1, chain_length=0):
    '''Moves a price series forward through n time steps using an analytical method.
     
     Details
     -------
     Price trajectories have constant drift terms and growth is reduced with noise.
     Returns a price series along with its corisponding wiener and time series.
     
     Argumets
     ---------
     n = the number of steps taken
     dt = the size of the time steps
     a = constant controling the significance of price drift
     b = constant controling the significance of the noise
     m = the default case refers to standard geometric brownian motion
     mu = mean of the normal distribution
     sigma = standard deviation of the normal distribution
     w_init = Initial value of the wiener series
     p_init = Initial value of the price series. If chaining, give p_init of the
         previously existing series which the returned one will be linked to.
     chain_length = The number of time steps in the previously existing series which
         the returned ones will be linked to. (default assumes no chaining)
     '''
    w_t = w_init
    p_t = p_init
    if chain_length>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
        trim = 0
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [float(w_init)]
        p_series = [float(p_init)]
        trim = 1
    for t in t_series[trim:]:
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        w_t += nu
        p_series.append(p_init*np.exp((a-0.5*b**2)*t + b*w_t))
        w_series.append(w_t)
    return t_series,w_series,p_series

def euler_retrace(series, x, y_init, dt, a, b, m=1, mu=0, sigma=1, chain=False):
    """Returns a price/wiener series based on a retrace of a wiener/price series.
    
    series = 'wiener'
    -----------------
    Retraces the values of the wiener series to create a price series.
    x = The retraced wiener series.
    y_init = Initial value of the price series.
    
    series = 'price':
    -----------------
    Retraces the values of the price series to create a wiener series.
    x = The retraced price series.
    y_init = Initial value of the weiner series.
    
    Parameters
    ----------
    dt = The size of the time steps
    a = Constant controling the significance of price drift
    b = Constant controling the significance of the noise
    m = The default case refers to standard geometric brownian motion
    mu = Mean of the normal distribution
    sigma = Standard deviation of the normal distribution
    chain = Return the series excluding its initial value?
    """
    
    if series == 'wiener':
        p_t = y_init
        w_series = x
        if chain==True:
            p_series = []
        else:
            p_series = [y_init]
        for i in range(1,len(w_series)):
            nu = w_series[i]-w_series[i-1]
            p_t += (a*dt+b*nu)*p_t**(m)
            p_series.append(p_t)
        return p_series
            
    elif series == 'price':
        w_t = y_init
        p_series = x
        if chain==True:
            w_series = []
        else:
            w_series = [y_init]
        for i in range(1,len(p_series)):
            dP = (p_series[i] - p_series[i-1])/(p_series[i-1]**(m))
            w_t+=(dP-a*dt)/b
            w_series.append(w_t)
        return w_series
    
def analy_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, p_init=1, chain=False):
    """Returns a price/wiener series based on a retrace of a wiener/price series.
    
    series = 'wiener'
    -----------------
    Retraces the values of the wiener series to create a price series.
    x = a wiener series.
    
    series = 'price':
    -----------------
    Retraces the values of the price series to create a wiener series.
    x = a price series.
    
    Parameters
    ----------
    dt = the size of the time steps.
    a = constant controling the significance of price drift.
    b = constant controling the significance of the noise.
    m = the default case refers to standard geometric brownian motion.
    mu = mean of the normal distribution.
    sigma = standard deviation of the normal distribution.
    p_init = Initial value of the price series.
    chain = Return the series excluding its initial value?
    """
    
    
    if series == 'wiener':
        p_series = []
        w_series = x
        if chain==True:
            i = 1
        else:
            i = 0
        for t in [dt*j for j in range(i,len(w_series))]:
            alph = (a-0.5*b**2)*t
            beta = b*w_series[i]
            p_series.append(p_init*np.exp(alph + beta))
            i+=1
        return p_series
    
    
    if series == 'price':
        p_series = x
        w_series = []
        if chain==True:
            i = 1
        else:
            i = 0
        for t in [dt*j for j in range(i,len(p_series))]:
            alph = (a-0.5*b**2)*t
            w_series.append((np.log(float(p_series[i])/p_init)-alph)/b)
            i+=1
        return w_series
```

##Analysis of the above Methods
+ ###I demonstrate that...
    1. A price series generated using a particular method can be recovered by retracing it with the same method. However this is only to a degree of approximation since there appears to be a problem with floating point precision. I say this because when there are only a minimal number of steps taken in a series there is a very high chance that the original and its retrace will be equivalent while this will almost certainly not be the case for series with a larger number of steps.
    2. Likewise, a wiener series generated using a particular method can be recovered by retracing the price series with the same method. Again, as above the floating point precision plays a role in the calculation and makes this true to a degree of approximation, but variances still remain relatively negligable.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('wiener',x)` can, to a reasonable degree be used interchangably. In the plots below a price series is generated with the euler/analytical method and then its corrisponding wiener series is retraced with an analytical/euler method. The variance between the two resuls is also displayed. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. Here a wiener series is generated with the euler/analytical method and then its corrisponding price series is retraced with an analytical/euler method. The variance between the two price retraces are shown to be similar to those in wiener retraces.

#Exponential Brownian Motion With Impetus

```python
def euler_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, **kwargs): 
    '''Moves the price series forward through the specified number of steps using Euler's method.'''
    
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    for i in range(number):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        w_series.append(w_series[-1]+nu)
        mu0 = (pval_crnt-pval_last)/pval_last

        pval_last = pval_crnt
        alph = a*mu0*dt
        beta = b*nu

        pval_crnt += (alph+beta)*pval_crnt**(m)
        p_series.append(pval_crnt)
        
    return w_series,p_series

def analy_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, **kwargs):
    '''Moves the price series forward through the specified number of steps using an analytical method.'''
    
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    if len(p_series)!=len(w_series):
        raise ValueError('w_series and p_series must be the same length')
    W_t = w_series[-1]
    
    for t in np.arange(len(w_series)*dt,(number+len(w_series))*dt,dt):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        W_t += nu

        alph = (dt*a*b*W_t-0.5*b**2)*t
        beta = b*W_t
        p_series.append(p_series[0]*np.exp((alph + beta)/(1-a*dt)**2))
        w_series.append(W_t)

    return w_series,p_series

def euler_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.


    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'weiner':
        p_series = [pval_last]
        w_series = x
        for i in range(1,len(w_series)):
            nu = w_series[i]-w_series[i-1]
            mu0 = (pval_crnt-pval_last)/pval_last
            pval_last = pval_crnt
            alph = a*mu0*dt
            beta = b*nu

            pval_crnt += (alph+beta)*pval_crnt**(m)
            p_series.append(pval_crnt)
        
        return p_series

    elif series == 'price':
        w_series = [0]
        p_series = x
        for i in range(1,len(p_series)):

            pval_next = p_series[i]

            dP = (pval_next - pval_crnt)/(pval_crnt**(m))
            mu0 = (pval_crnt - pval_last)/pval_last
            w_series.append(w_series[-1]+(dP-a*mu0*dt)/b)

            pval_last = pval_crnt
            pval_crnt = pval_next
        
        return w_series

def analy_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_init=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.

    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'weiner':
        p_series = [pval_init]
        w_series = x
        
        i = 1
        for t in [dt*j for j in range(1,len(w_series))]:
            alph = (dt*a*b*w_series[i]-0.5*b**2)*t
            beta = b*w_series[i]
            p_series.append(p_series[0]*np.exp((alph + beta)/(1-a*dt)**2))

            i+=1
        
        return p_series

    if series == 'price':
        w_series = [0]
        p_series = x
        
        i = 1
        mu0 = 0
        for t in [dt*j for j in range(1,len(p_series))]:
            alph = 1-a*dt
            beta = b*(1-t*a*dt)
            w_series.append((np.log(float(p_series[i])/p_series[0])*alph**2+0.5*b**2*t)/beta)
            i+=1
            
        return w_series
```

    1. A price series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.
    2. Likewise, a wiener series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.

A notable point to make is that as the time steps a changed to be smaller, the two methods will become more similar and variance will decrease drastically. Thus strongly suggests that the two are indeed replicating the same process. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. And again, as time steps become smaller the more similar the two methods become.
