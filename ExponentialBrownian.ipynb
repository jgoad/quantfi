#Exponential Brownian Motion With Drift

```python
def euler_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, **kwargs): 
    '''Moves the price series forward through the specified number of steps using Euler's method.'''
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    
    for i in range(number):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        w_series.append(w_series[-1]+nu)

        pval_last = pval_crnt
        alph = a*dt
        beta = b*nu

        pval_crnt += (alph+beta)*pval_crnt**(m)
        p_series.append(pval_crnt)
    
    return w_series, p_series

def analy_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, w_series=[0], **kwargs):
    '''Moves the price series forward through the specified number of steps using an analytical method.'''
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    if len(p_series)!=len(w_series):
        raise ValueError('w_series and p_series must be the same length')
    
    W_t = w_series[-1]
    mu0 = (pval_crnt-pval_last)/pval_last
    
    for t in np.arange(len(p_series)*dt,(number+len(p_series))*dt,dt):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        W_t += nu

        alph = (a-0.5*b**2)*t
        beta = b*W_t
        p_series.append(p_series[0]*np.exp(alph + beta))
        w_series.append(W_t)

        mu0 = (p_series[-1]-p_series[-2])/p_series[-2]
        
    return w_series, p_series

def euler_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.

    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'weiner':
        
        w_series = x
        p_series = [pval_crnt]

        for i in range(1,len(w_series)):
            nu = w_series[i]-w_series[i-1]
            pval_last = pval_crnt
            alph = a*dt
            beta = b*nu

            pval_crnt += (alph+beta)*pval_crnt**(m)
            p_series.append(pval_crnt)
        
        return p_series
            
    elif series == 'price':
        p_series = x
        w_series = [0]

        for i in range(1,len(p_series)):

            pval_next = p_series[i]

            dP = (pval_next - pval_crnt)/(pval_crnt**(m))
            w_series.append(w_series[-1]+(dP-a*dt)/b)

            pval_last = pval_crnt
            pval_crnt = pval_next
        
        return w_series
    
def analy_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_init=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.

    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''
    
    p_series = [pval_init]
    if series == 'weiner':
        w_series = x
        i = 1
        mu0 = 0
        for t in [dt*j for j in range(1,len(w_series))]:
            alph = (a-0.5*b**2)*t
            beta = b*w_series[i]
            p_series.append(p_series[0]*np.exp(alph + beta))
            i+=1
        
        return p_series
    
    w_series = [0]
    if series == 'price':
        p_series = x
        i = 1
        w_series = [0]
        for t in [dt*j for j in range(1,len(p_series))]:
            alph = (a-0.5*b**2)*t
            w_series.append((np.log(float(p_series[i])/pval_init)-alph)/b)
            i+=1
            
        return w_series
```

##Analysis of the above Methods
+ ###I demonstrate that...
    1. A price series generated using a particular method can be recovered by retracing it with the same method. However this is only to a degree of approximation since there appears to be a problem with floating point precision. I say this because when there are only a minimal number of steps taken in a series there is a very high chance that the original and its retrace will be equivalent while this will almost certainly not be the case for series with a larger number of steps.
    2. Likewise, a weiner series generated using a particular method can be recovered by retracing the price series with the same method. Again, as above the floating point precision plays a role in the calculation and makes this true to a degree of approximation, but variances still remain relatively negligable.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('weiner',x)` can, to a reasonable degree be used interchangably. In the plots below a price series is generated with the euler/analytical method and then its corrisponding weiner series is retraced with an analytical/euler method. The variance between the two resuls is also displayed. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. Here a weiner series is generated with the euler/analytical method and then its corrisponding price series is retraced with an analytical/euler method. The variance between the two price retraces are shown to be similar to those in weiner retraces.

```python
number = 5
params = [0.001,1,1]

W1,P1 = euler_step(number,0.001,1,1)
P2 = euler_retrace('weiner',W1,*params)

print "euler_step == euler_retrace('weiner',x) for series with {0} steps:".format(number), P1 == P2
print 'average variance:', np.average(abs(np.array(P1)-np.array(P2)))

print

W1,P1 = analy_step(number,0.001,1,1)
P2 = analy_retrace('weiner',W1,*params)

print "analy_step == analy_retrace('weiner',x) for series with {0} steps:".format(number), P1 == P2
print 'average variance:', np.average(abs(np.array(P1)-np.array(P2)))
```

```python
def euler_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, **kwargs): 
    '''Moves the price series forward through the specified number of steps using Euler's method.'''
    
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    for i in range(number):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        w_series.append(w_series[-1]+nu)
        mu0 = (pval_crnt-pval_last)/pval_last

        pval_last = pval_crnt
        alph = a*mu0*dt
        beta = b*nu

        pval_crnt += (alph+beta)*pval_crnt**(m)
        p_series.append(pval_crnt)
        
    return w_series,p_series

def analy_step(number, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1, **kwargs):
    '''Moves the price series forward through the specified number of steps using an analytical method.'''
    
    w_series = kwargs.get('w_series',[0])
    p_series = kwargs.get('p_series',[pval_crnt])
    if len(p_series)!=len(w_series):
        raise ValueError('w_series and p_series must be the same length')
    W_t = w_series[-1]
    
    for t in np.arange(len(w_series)*dt,(number+len(w_series))*dt,dt):
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        W_t += nu

        alph = (dt*a*b*W_t-0.5*b**2)*t
        beta = b*W_t
        p_series.append(p_series[0]*np.exp((alph + beta)/(1-a*dt)**2))
        w_series.append(W_t)

    return w_series,p_series

def euler_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_last=1, pval_crnt=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.


    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'weiner':
        p_series = [pval_last]
        w_series = x
        for i in range(1,len(w_series)):
            nu = w_series[i]-w_series[i-1]
            mu0 = (pval_crnt-pval_last)/pval_last
            pval_last = pval_crnt
            alph = a*mu0*dt
            beta = b*nu

            pval_crnt += (alph+beta)*pval_crnt**(m)
            p_series.append(pval_crnt)
        
        return p_series

    elif series == 'price':
        w_series = [0]
        p_series = x
        for i in range(1,len(p_series)):

            pval_next = p_series[i]

            dP = (pval_next - pval_crnt)/(pval_crnt**(m))
            mu0 = (pval_crnt - pval_last)/pval_last
            w_series.append(w_series[-1]+(dP-a*mu0*dt)/b)

            pval_last = pval_crnt
            pval_crnt = pval_next
        
        return w_series

def analy_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, pval_init=1):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.

    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'weiner':
        p_series = [pval_init]
        w_series = x
        
        i = 1
        for t in [dt*j for j in range(1,len(w_series))]:
            alph = (dt*a*b*w_series[i]-0.5*b**2)*t
            beta = b*w_series[i]
            p_series.append(p_series[0]*np.exp((alph + beta)/(1-a*dt)**2))

            i+=1
        
        return p_series

    if series == 'price':
        w_series = [0]
        p_series = x
        
        i = 1
        mu0 = 0
        for t in [dt*j for j in range(1,len(p_series))]:
            alph = 1-a*dt
            beta = b*(1-t*a*dt)
            w_series.append((np.log(float(p_series[i])/p_series[0])*alph**2+0.5*b**2*t)/beta)
            i+=1
            
        return w_series
```

##Analysis of the above Methods
+ ###Similarly to before I demonstrate that...
    1. A price series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.
    2. Likewise, a weiner series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('weiner',x)` can, to a reasonable degree be used interchangably. A notable point to make is that as the time steps a changed to be smaller, the two methods will become more similar and variance will decrease drastically. Thus strongly suggests that the two are indeed replicating the same process. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. And again, as time steps become smaller the more similar the two methods become.
