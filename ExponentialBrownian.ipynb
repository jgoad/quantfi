```python
def wiener(n, dt, t_init=0, w_init=0.0):
    """Returns one realization of a Wiener process with n steps of length dt.
    The time and Wiener series can be initialized using t_init and w_init respectively."""
    n+=1
    t_series = np.arange(t_init,n*dt,dt)
    h = t_series[1]-t_series[0]
    z = np.random.normal(0.0,1.0,n)
    dw = np.sqrt(h)*z
    dw[0] = w_init
    w_series = dw.cumsum()
    return t_series, w_series
```

#Exponential Brownian Motion With Drift

```python
def euler_step(n, dt, a, b, m=1, mu=0, sigma=1, w_init=0, p_init=1, chain_length=0): 
    '''Moves a price series forward through n time steps using Euler's method.
    
    Details
    -------
    The Stochastic DE has conatins drift term and growth is reduced by noise.
    Returns a price series along with its corisponding wiener and time series.
    
    Argumets
    --------
    n = The number of steps taken
    dt = The size of the time steps
    a = Constant controling the significance of price drift
    b = Constant controling the significance of the noise
    m = The default case refers to standard geometric brownian motion
    mu = Mean of the normal distribution
    sigma = Standard deviation of the normal distribution
    w_init = Initial value of the wiener series
    p_init = Initial value of the price series. If chaining, give p_init of the
        previously existing series which the returned one will be linked to.
    chain_length = The number of time steps in the previously existing series which
        the returned ones will be linked to. (default assumes no chaining)
    
    (!) Note: it is suggested that all arguments be given as floats.
    '''
    w_t = w_init
    p_t = p_init
    if chain_length>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [w_init]
        p_series = [p_init]
    for i in range(n):
        dw = random.gauss(mu,np.sqrt(dt)*sigma)
        # Incrament p(t) with dp as given by the DE.
        # Incrament w(t) with dw as given by the normal distribution
        p_t += (a*dt+b*dw)*p_t**(m)
        w_t += dw
        # Append new values to each series;
        w_series.append(w_t)
        p_series.append(p_t)
    return t_series,w_series,p_series

def analy_step(n, dt, a, b, m=1, mu=0, sigma=1, w_init=0, p_init=1, chain_length=0):
    '''Moves a price series forward through n time steps using an analytical method.
    
    Details
    -------
    The Stochastic DE has conatins drift term and growth is reduced by noise.
    Returns a price series along with its corisponding wiener and time series.
    
    Argumets
    ---------
    n = The number of steps taken
    dt = The size of the time steps
    a = Constant controling the significance of price drift
    b = Constant controling the significance of the noise
    m = The default case refers to standard geometric brownian motion
    mu = Mean of the normal distribution
    sigma = Standard deviation of the normal distribution
    w_init = Initial value of the wiener series
    p_init = Initial value of the price series. If chaining, give p_init of the
        previously existing series which the returned one will be linked to.
    chain_length = The length of the previously existing price series
        (default assumes no chaining)
    
    (!) Note: it is suggested that all arguments be given as floats.
    '''
    w_t = w_init
    p_t = p_init
    if chain_length>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
        trim = 0
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [float(w_init)]
        p_series = [float(p_init)]
        trim = 1
    for t in t_series[trim:]:
        dw = random.gauss(mu,np.sqrt(dt)*sigma)
        w_t += dw
        p_series.append(p_init*np.exp((a-0.5*b**2)*t + b*w_t))
        w_series.append(w_t)
    return t_series,w_series,p_series

def euler_retrace(series, x, y_init, dt, a, b, m=1, mu=0, sigma=1, chain=False):
    """Returns a price/wiener series based on a retrace of a wiener/price series.
    
    series = 'wiener'
    -----------------
    Retraces the values of the wiener series to create a price series.
    x = The retraced wiener series.
    y_init = Initial value of the price series.
    
    series = 'price':
    -----------------
    Retraces the values of the price series to create a wiener series.
    x = The retraced price series.
    y_init = Initial value of the weiner series.
    
    Parameters
    ----------
    dt = The size of the time steps
    a = Constant controling the significance of price drift
    b = Constant controling the significance of the noise
    m = The default case refers to standard geometric brownian motion
    mu = Mean of the normal distribution
    sigma = Standard deviation of the normal distribution
    chain = Return the series excluding its initial value?
    
    (!) Note: it is suggested that all arguments be given as floats.
    """
    
    if series == 'wiener':
        p_t = y_init
        w_series = x
        if chain==True:
            p_series = []
        else:
            p_series = [y_init]
        for i in range(1,len(w_series)):
            dw = w_series[i]-w_series[i-1]
            p_t += (a*dt+b*dw)*p_t**(m)
            p_series.append(p_t)
        return p_series
            
    elif series == 'price':
        w_t = y_init
        p_series = x
        if chain==True:
            w_series = []
        else:
            w_series = [y_init]
        for i in range(1,len(p_series)):
            dP = (p_series[i] - p_series[i-1])/(p_series[i-1]**(m))
            w_t+=(dP-a*dt)/b
            w_series.append(w_t)
        return w_series
    
def analy_retrace(series, x, dt, a, b, m=1, mu=0, sigma=1, p_init=1, chain=False):
    """Returns a price/wiener series based on a retrace of a wiener/price series.
    
    series = 'wiener'
    -----------------
    Retraces the values of the wiener series to create a price series.
    x = a wiener series.
    
    series = 'price':
    -----------------
    Retraces the values of the price series to create a wiener series.
    x = a price series.
    
    Parameters
    ----------
    dt = The size of the time steps.
    a = Constant controling the significance of price drift.
    b = Constant controling the significance of the noise.
    m = The default case refers to standard geometric brownian motion.
    mu = Mean of the normal distribution.
    sigma = Standard deviation of the normal distribution.
    p_init = Initial value of the price series.
    chain = Return the series excluding its initial value?
    
    (!) Note: it is suggested that all arguments be given as floats.
    """
    
    
    if series == 'wiener':
        p_series = []
        w_series = x
        if chain==True:
            i = 1
        else:
            i = 0
        for t in [dt*j for j in range(i,len(w_series))]:
            alph = (a-0.5*b**2)*t
            beta = b*w_series[i]
            p_series.append(p_init*np.exp(alph + beta))
            i+=1
        return p_series
    
    
    if series == 'price':
        p_series = x
        w_series = []
        if chain==True:
            i = 1
        else:
            i = 0
        for t in [dt*j for j in range(i,len(p_series))]:
            alph = (a-0.5*b**2)*t
            w_series.append((np.log(float(p_series[i])/p_init)-alph)/b)
            i+=1
        return w_series
```

##Analysis of the above Methods
+ ###I demonstrate that...
    1. A price series generated using a particular method can be recovered by retracing it with the same method. However this is only to a degree of approximation since there appears to be a problem with floating point precision. I say this because when there are only a minimal number of steps taken in a series there is a very high chance that the original and its retrace will be equivalent while this will almost certainly not be the case for series with a larger number of steps.
    2. Likewise, a wiener series generated using a particular method can be recovered by retracing the price series with the same method. Again, as above the floating point precision plays a role in the calculation and makes this true to a degree of approximation, but variances still remain relatively negligable.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('wiener',x)` can, to a reasonable degree be used interchangably. In the plots below a price series is generated with the euler/analytical method and then its corrisponding wiener series is retraced with an analytical/euler method. The variance between the two resuls is also displayed. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. Here a wiener series is generated with the euler/analytical method and then its corrisponding price series is retraced with an analytical/euler method. The variance between the two price retraces are shown to be similar to those in wiener retraces.

#Exponential Brownian Motion With Drift and Momentum

---

###Numerical solution implamented in euler_step and euler_retrace

$$
\Delta X_j = X_{j+1}-X_j \text{ and } \Delta X_{j-1} = X_j-X_{j-1}
$$

$$
\frac{\Delta X_j}{X_j^m} = \alpha_1\Delta t+ \alpha_2 \frac{\Delta X_{j-1}}{X_{j-1}^m}\Delta t + \beta\Delta W
$$

###Analytical solution impamented in analy_step and analy_retrace
 
Because $\Delta X_j/X_j\approx\Delta X_{j-1}/X_{j-1}$ for $\Delta t$ sufficiently small (i.e. $\Delta t \rightarrow h)$, we can write the continuous differential equation as

$$
\frac{\Delta X_j}{X_j} = \alpha_1 h+ \alpha_2 \frac{\Delta X_j}{X_j} h + \beta\Delta W
$$

$$
\frac{\Delta X}{X} = \frac{\alpha_1 h + \beta\Delta W}{1-\alpha_2 h}.
$$

Then using a difference of logs to solve as detailed on page 40 of [A. J. Roberts' textbook](http://tinyurl.com/AJRoberts-Calculus-of-Finance) on the calculus of financial mathematics, the solution is

$$
X = X_0\exp{\bigg[\frac{(1-h\alpha_1 - h\alpha_2)(\alpha_1 t_n + \beta W(t_n)) - \frac{1}{2}(\beta^2-\alpha_1^2 h)t_n}{(1-\alpha_2 h)^2}\bigg]}.
$$

Which can be generalized to the form

$$
X = X_0\exp{\bigg[\frac{at + bW(t)}{c}\bigg]}.
$$

At which point, applying Ito's formula to this expression, where we assert that $c = (1-\alpha_2 h)$, we find

$$
\frac{dX}{X} = \frac{(a+\frac{1}{2c}b^2)dt + bdW}{c},
$$

Which based on the similarity between this and the initial differential equation implies that $\alpha_1 = (a+\frac{1}{2c}b^2)$ and $\beta = b$. Thus verifying the analytical solution. With this result we can rewrite the analytical solution in this form,

$$
X = X_0\exp{\bigg[\frac{(\alpha_1 - \frac{1}{2c}\beta^2)t+\beta W(t)}{c}\bigg]}.
$$

```python
def euler_step(n, dt, a, b, c, m=1, mu=0, sigma=1, w_init=0., p_init=1., chain=(0,1)): 
    '''Moves a price series forward through n time steps using Euler's method.
    
    Details
    -------
    The Stochastic DE has drift and momentum terms where growth is reduced by noise.
    Returns a price series along with its corisponding wiener and time series.
    
    Argumets
    --------
    n = The number of steps taken.
    dt = The size of the time steps.
    a = Constant controling the significance of price drift.
    b = Constant controling the significance of the noise.
    c = Constant controling the influence of momentum (c=0 implies no influence).
    m = The default case refers to standard geometric brownian motion.
    mu = Mean of the normal distribution.
    sigma = Standard deviation of the normal distribution.
    w_init = Initial value of the wiener series.
    p_init = Initial value of the price series. If chaining, give p_init of the
        previously existing series which the returned one will be linked to.
    chain = Tuple with length of previously existing price series and ending value.
    
    (!) Note: it is suggested that all arguments be given as floats.
    '''
    w_t = w_init
    p_t = p_init
    p_last = p_init
    if chain[0]>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [w_init]
        p_series = [p_init]
    for i in range(n):
        dw = random.gauss(mu,np.sqrt(dt)*sigma)
        dp = (p_t-p_last)/p_last
        p_last = p_t
        #incrament p(t) and w(t), then append values
        w_t += dw
        p_t += (a*dt + c*dp*dt + b*dw)*p_t**(m)
        w_series.append(w_t)
        p_series.append(p_t)
    return t_series,w_series,p_series

def analy_step(n, dt, a, b, c, mu=0, sigma=1, w_init=0., p_init=1., chain_length=0):
    '''Moves a price series forward through n time steps using an analytical method.
    
    Details
    -------
    The Stochastic DE has drift and momentum terms where growth is reduced by noise.
    Returns a price series along with its corisponding wiener and time series.
    Analagous only to the case where m=1.
    
    Argumets
    --------
    n = The number of steps taken.
    dt = The size of the time steps.
    a = Constant controling the significance of price drift.
    b = Constant controling the significance of the noise.
    c = Constant controling the influence of momentum (c!=1, c=0 => no influence).
    m = The default case refers to standard geometric brownian motion.
    mu = Mean of the normal distribution.
    sigma = Standard deviation of the normal distribution.
    w_init = Initial value of the wiener series.
    p_init = Initial value of the price series. If chaining, give p_init of the
        previously existing series which the returned one will be linked to.
    chain = Tuple with length of previously existing price series and ending value.
    
    (!) Note: it is suggested that all arguments be given as floats.
    '''
    
    w_t = w_init
    p_t = p_init
    if chain_length>0:
        t_series = np.arange(chain_length*dt,(n+chain_length-0.9)*dt,dt)
        w_series = []
        p_series = []
        trim = 0
    else:
        t_series = np.arange(0,(n+0.1)*dt,dt)
        w_series = [float(w_init)]
        p_series = [float(p_init)]
        trim = 1
    for t in t_series[trim:]:
        nu = random.gauss(mu,np.sqrt(dt)*sigma)
        w_t += nu
        p_series.append(p_init*np.exp(((a-b**2/2./(1-c*dt))*t+b*w_t)/(1-c*dt)))
        w_series.append(w_t)
    return t_series,w_series,p_series

def euler_retrace(series, x, y_init, dt, a, b, c, m=1, mu=0, sigma=1, y_last=None, chain=False):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.


    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''

    if series == 'wiener':
        w_series = x
        p_t = y_init
        p_last = y_init
        if chain==True:
            p_series = []
        else:
            p_series = [float(y_init)]
        for i in range(1,len(w_series)):
            dw = w_series[i]-w_series[i-1]
            dp = (p_t-p_last)/p_last
            p_last = p_t
            #incrament p(t) and w(t), then append values
            p_t += (a*dt + c*dp*dt + b*dw)*p_t**(m)
            p_series.append(p_t)
        
        return p_series

    elif series == 'price':
        if y_last == None:
            if chain == True:
                raise TypeError('must input a value for y_last')
            else:
                y_last = 1
        w_t = y_init
        p_series = [y_last] + x
        if chain==True:
            w_series = []
        else:
            w_series = [float(y_init)]
        for i in range(2,len(p_series)):
            dp0 = (p_series[i-1] - p_series[i-2])/p_series[i-2]
            dp1 = (p_series[i] - p_series[i-1])
            w_t+=(dp1/p_series[i-1]**(m) - a*dt - c*dp0*dt)/b
            w_series.append(w_t)
        return w_series

def analy_retrace(series, x, dt, a, b, c, mu=0, sigma=1, p_init=1, chain=False):
    '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.

    series = 'weiner': 
    Retraces the values of the weiner series to create a price series.
    series = 'price':
    Retraces the values of the price series to create a weiner series.

    If x != None then x is overrides preexisting series data and is
    considered to be the series which is being retraced.
    NOTE: the provided series must have more than one value.'''
    
    if series == 'wiener':
        p_series = []
        w_series = x
        if chain==True:
            i = 1
        else:
            i = 0
        for t in [dt*j for j in range(i,len(w_series))]:
            p_series.append(p_init*np.exp(((a-b**2/2./(1-c*dt))*t+b*w_series[i])/(1-c*dt)))
            i+=1
        return p_series

    if series == 'price':
        w_series = [0.]
        p_series = x
        i = 1
        mu0 = 0
        for t in [dt*j for j in range(1,len(p_series))]:
            w_series.append((np.log(p_series[i]/p_init)*(1-c*dt)-(a-b**2/2./(1-c*dt))*t)/b)
            i+=1
        return w_series
```

##Analysis of the above Methods
+ ###Similarly to before I demonstrate that...
    1. A price series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.
    2. Likewise, a wiener series generated using a particular method can be recovered to a degree of approximation by retracing it with the same method.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('wiener',x)` can, to a reasonable degree be used interchangably. A significant point to make is that as the initial assumption of small time steps would suggest, the two methods will become more similar and variance will decrease drastically the smaller the time steps are.
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. Though it is not displayed, it is again true that as the time steps become smaller the more similar the two methods become.
