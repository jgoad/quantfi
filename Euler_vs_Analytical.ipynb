```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

#Examining solutions to Ito SDE's using Euler and analytical solutions
+ Eventually this variation on `exponential_brownian` below will have a plotting method implamented which is similar to `plot_history` in `ExpBrownian.ipynb`. At such a point the following code can replace what's seens above due to the fact that it includes all same capabilities in addition to many others.

```python
class exponential_brownian:
    '''Generates data using equation (1) from Sornette paper.
    m=1 is supposed to be equivalent to brownian motion
    '''
    
    class FillError(Exception):
        def __init__(self, kind):
            self.string = ('The {0} series has not been defined. '
                           "Pass it one in the class constructor through **kwargs or arg 'x' in retrace()").format(kind)
        def __str__(self):
            return str(self.string)
    
    def __init__(self, dt, m, a, b, mu=0, sigma=1, **kwargs):
        '''Provide the parameters included in the differential equation above.
        **kwargs makes it possible to initialize with a preexisting price or weiner series.'''
        
        self.dt = dt
        self.m = m
        self.a = a
        self.b = b
        self.mu = mu
        self.sigma = sigma
        
        self.init_pval_last = kwargs.get('pval_last',1.0)
        self.init_pval_crnt = kwargs.get('pval_crnt',1.0)
        
        self.pval_last = self.init_pval_last
        self.pval_crnt = self.init_pval_crnt
        self.p_series = kwargs.get('p_series',[self.pval_crnt])
        self.w_series = kwargs.get('w_series',[0])
        self.t_steps = 1
    
    def euler_step(self, number): 
        '''Moves the price series forward through the specified number of steps using Euler's method.'''
        for i in range(number):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            self.w_series.append(self.w_series[-1]+nu)
            mu0 = (self.pval_crnt-self.pval_last)/self.pval_last

            self.pval_last = self.pval_crnt
            alph = self.a*mu0*self.dt
            beta = self.b*nu

            self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
            self.p_series.append(self.pval_crnt)
            self.t_steps+=1
    
    def analy_step(self, number):
        '''Moves the price series forward through the specified number of steps using an analytical method.'''
        W_t = self.w_series[-1]
        mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
        for i in np.arange(self.t_steps*self.dt,(number+self.t_steps)*self.dt,self.dt):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            W_t += nu

            alph = (self.a*mu0-0.5*self.b**2)*i
            beta = self.b*W_t
            self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))
            self.w_series.append(W_t)

            mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
            self.t_steps+=1

        self.pval_last = self.p_series[-2]
        self.pval_crnt = self.p_series[-1]
            
    def euler_retrace(self, series, x=None):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            self.pval_last = self.init_pval_last
            self.pval_crnt = self.init_pval_crnt
            self.p_series = [self.pval_last]
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            for nu in self.w_series[1:]:
                mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
                self.pval_last = self.pval_crnt
                alph = self.a*mu0*self.dt
                beta = self.b*nu*np.sqrt(self.dt)

                self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
                self.p_series.append(self.pval_crnt)

        elif series == 'price':
            self.w_series = [0]

            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)

            pval_last = self.init_pval_last
            pval_crnt = self.init_pval_crnt

            for i in range(1,len(self.p_series)):

                pval_next = self.p_series[i]

                dP = (pval_next - pval_crnt)/(pval_crnt**(self.m))
                mu0 = (pval_crnt - pval_last)/pval_last
                self.w_series.append(self.w_series[-1]+(dP-self.a*mu0*self.dt)/self.b)

                pval_last = pval_crnt
                pval_crnt = pval_next
                    
    def analy_retrace(self, series, x=None):
        if series == 'weiner':
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            count = 1
            mu0 = 0
            for i in [self.dt*i for i in range(1,len(self.w_series))]:
                alph = (self.a*mu0-0.5*self.b**2)*i
                beta = self.b*self.w_series[count]
                self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))

                mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
                count+=1
        
        if series == 'price':
            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)
```

###Demonstration of new `exponential_brownian` class
+ Here I demonstrate that the `euler` and `analytical` methods from `step()` and `retrace()` can be used interchangably with the exception of a caviot showcased in Figures (2) and (3). This is given more thought in the markdown below.

```python
Stock1 = exponential_brownian(0.001,1,1,0.1)
Stock1.euler_step(200)

Stock2 = exponential_brownian(0.001,1,1,0.1)
Stock2.analy_retrace('weiner', Stock1.w_series)

#-------------------------

x1=range(len(Stock1.w_series))
x2=range(len(Stock2.w_series))

#x1=Stock1.w_series
#x2=Stock2.w_series

#y1=Stock1.w_series
#y2=Stock2.w_series

y1=Stock1.p_series
y2=Stock2.p_series

plt.figure(figsize = (10,5))
plt.plot(x1,y1,color='b')
plt.plot(x2,y2,color='r')
plt.title('Fig(1): Plot of y1 and y2.', fontsize=14)
plt.show()

plt.figure(figsize = (10,5))
plt.plot(x1,abs(np.array(y1)-np.array(y2)))
plt.title('Fig(2): Variance between y1 and y2.', fontsize=14)
plt.show()

#-----------------------------------------

Stock1 = exponential_brownian(0.001,1,1,0.1)
Stock1.analy_step(1060)

x1=range(len(Stock1.w_series))
y1=Stock1.p_series

plt.figure(figsize = (10,5))
plt.plot(x1,y1,color='b')

plt.xlim(0,1080)
plt.title('Fig(3): Extension of a seperate analytical price series.', fontsize=14)
plt.show()
```

+ In Figure (2) it's quite clear that as time goes on the analytical `retrace()` of a price series produced using the `euler` method begins to diverge. This might be expected as they are different methods after all, however I believe that Figure (3) shows where this problem actually originates.
+ It may be that I have not implameted the analytical solution correctly, but when I considered the solution given by equation 2.2 on page 40 in [A. J. Roberts' textbook](http://tinyurl.com/aj-roberts-txtbook) it seemed more likely that it came down to the fact that since $W(t)$ only grows like $\sqrt{t}$, the dominant term in the argument of the exponential function comes from $t$. In the book it discusses this and mentions that in the case where $\alpha$ is constant $X(t)$ eventually goes to zero. However in the particular case I have here, $\alpha = a \times \mu_0$ where $\mu_0$ is equal to the instantaneous return rate. Thus as $t$ increases so does the influence of $\mu0$ which is strongly controled by variations in $W(t)$.

```python
class analy_restored(exponential_brownian):
    def analy_step(self, number):
        '''Moves the price series forward through the specified number of steps using an analytical method.'''
        W_t = self.w_series[-1]
        mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
        for i in np.arange(self.t_steps*self.dt,(number+self.t_steps)*self.dt,self.dt):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            W_t += nu
            
            alph = (mu0+self.a-0.5*self.b**2)*i
            beta = self.b*W_t
            self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))
            self.w_series.append(W_t)

            mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
            self.t_steps+=1

        self.pval_last = self.p_series[-2]
        self.pval_crnt = self.p_series[-1]
```

```python
Stock1 = analy_restored(0.001,1,1,0.1)
Stock1.analy_step(1100)

x1=range(len(Stock1.w_series))
y1=Stock1.p_series

plt.figure(figsize = (10,5))
plt.plot(x1,y1,color='b')

plt.title('Fig(3): Extension of a seperate analytical price series.', fontsize=14)
plt.show()
```
