#Examining solutions to Ito SDE's using the Euler and analytical methods
+ Eventually this variation on `exponential_brownian` from `ExpBrownian.ipynb` will have a plotting method implamented which is similar to `plot_history` in the same file. At such a point the following code can replace what's seens there due to the fact that it includes all same capabilities in addition to many others.

```python
class exp_brownian:
    
    class FillError(Exception):
        def __init__(self, kind):
            self.string = ('The {0} series has not been defined. '
                           "Pass it one in the class constructor through **kwargs or arg 'x' in retrace()").format(kind)
        def __str__(self):
            return str(self.string)
    
    def __init__(self, dt, m, a, b, mu=0, sigma=1, **kwargs):
        '''Provide the parameters included in the differential equation above.
        **kwargs makes it possible to initialize with a preexisting price or weiner series.'''
        
        self.dt = dt
        self.m = m
        self.a = a
        self.b = b
        self.mu = mu
        self.sigma = sigma
        
        self.init_pval_last = kwargs.get('pval_last',1.0)
        self.init_pval_crnt = kwargs.get('pval_crnt',1.0)
        
        self.pval_last = self.init_pval_last
        self.pval_crnt = self.init_pval_crnt
        self.p_series = kwargs.get('p_series',[self.pval_crnt])
        self.w_series = kwargs.get('w_series',[0])
        self.t_steps = 1
    
    def euler_step(self, number): 
        '''Moves the price series forward through the specified number of steps using Euler's method.'''
        for i in range(number):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            self.w_series.append(self.w_series[-1]+nu)

            self.pval_last = self.pval_crnt
            alph = self.a*self.dt
            beta = self.b*nu

            self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
            self.p_series.append(self.pval_crnt)
            self.t_steps+=1
    
    def analy_step(self, number):
        '''Moves the price series forward through the specified number of steps using an analytical method.'''
        W_t = self.w_series[-1]
        mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
        for i in np.arange(self.t_steps*self.dt,(number+self.t_steps)*self.dt,self.dt):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            W_t += nu
            
            alph = (self.a-0.5*self.b**2)*i
            beta = self.b*W_t
            self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))
            self.w_series.append(W_t)

            mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
            self.t_steps+=1

        self.pval_last = self.p_series[-2]
        self.pval_crnt = self.p_series[-1]
    
    def euler_retrace(self, series, x=None):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            self.pval_last = self.init_pval_last
            self.pval_crnt = self.init_pval_crnt
            self.p_series = [self.pval_last]
            
            for i in range(1,len(self.w_series)):
                nu = self.w_series[i]-self.w_series[i-1]
                self.pval_last = self.pval_crnt
                alph = self.a*self.dt
                beta = self.b*nu
                
                self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
                self.p_series.append(self.pval_crnt)
        
        elif series == 'price':
            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)
            self.w_series = [0]

            pval_last = self.init_pval_last
            pval_crnt = self.init_pval_crnt

            for i in range(1,len(self.p_series)):

                pval_next = self.p_series[i]

                dP = (pval_next - pval_crnt)/(pval_crnt**(self.m))
                self.w_series.append(self.w_series[-1]+(dP-self.a*self.dt)/self.b)

                pval_last = pval_crnt
                pval_crnt = pval_next
    
    def analy_retrace(self, series, x=None):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            i = 1
            mu0 = 0
            for t in [self.dt*j for j in range(1,len(self.w_series))]:
                alph = (self.a-0.5*self.b**2)*t
                beta = self.b*self.w_series[i]
                self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))
                
                i+=1
                
        if series == 'price':
            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)
            i = 1
            self.w_series = [0]
            for t in [self.dt*j for j in range(1,len(self.p_series))]:
                alph = (self.a-0.5*self.b**2)*t
                self.w_series.append((np.log(float(self.p_series[i])/self.init_pval_crnt)-alph)/self.b)
                i+=1
            
```

##details of `exp_brownian` class
+ ###I demonstrate that...
    1. A price series generated using a particular method can be recovered by retracing it with the same method. However this is only to a degree of approximation since there appears to be a problem with floating point precision. I say this because when there are only a minimal number of steps taken in a series there is a very high chance that the original and its retrace will be equivalent while this will almost certainly not be the case for series with a larger number of steps.
    2. Likewise, a weiner series generated using a particular method can be recovered by retracing the price series with the same method. Again, as above the floating point precision plays a role in the calculation and makes this true to a degree of approximation, but variances still remain relatively negligable.
    3. The methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('weiner',x)` can, to a reasonable degree be used interchangably. In the plots below a price series is generated with the euler/analytical method and then its corrisponding weiner series is retraced with an analytical/euler method. The variance between the two resuls is also displayed. 
    4. Similarly, the methods for `<euler/analytical>_step()` and `<euler/analytical>_retrace('price',x)` can, to a reasonable degree be used interchangably. Here a weiner series is generated with the euler/analytical method and then its corrisponding priec series is retraced with an analytical/euler method. The variance between the two `'price'` retraces are shown to be similar to those in `'weiner'` retraces.

#Examining the exponential brownian method with the inclusion of "momentum" with Euler and analytical methods

```python
class exp_mom_brownian(exp_brownian):
    '''Generates data using equation (1) from Sornette paper where mu_0 is the instantaneous return rate.
    In other words mu_0 as the instantaneous return rate adds momentum characteristics.
    '''
    
    def euler_step(self, number): 
        '''Moves the price series forward through the specified number of steps using Euler's method.'''
        for i in range(number):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            self.w_series.append(self.w_series[-1]+nu)
            mu0 = (self.pval_crnt-self.pval_last)/self.pval_last

            self.pval_last = self.pval_crnt
            alph = self.a*mu0*self.dt
            beta = self.b*nu

            self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
            self.p_series.append(self.pval_crnt)
            self.t_steps+=1
    
    def analy_step(self, number):
        '''Moves the price series forward through the specified number of steps using an analytical method.'''
        W_t = self.w_series[-1]
        mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
        for i in np.arange(self.t_steps*self.dt,(number+self.t_steps)*self.dt,self.dt):
            nu = random.gauss(self.mu,np.sqrt(self.dt)*self.sigma)
            W_t += nu

            alph = (self.a*mu0-0.5*self.b**2)*i
            beta = self.b*W_t
            self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))
            self.w_series.append(W_t)

            mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
            self.t_steps+=1

        self.pval_last = self.p_series[-2]
        self.pval_crnt = self.p_series[-1]
            
    def euler_retrace(self, series, x=None):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            self.pval_last = self.init_pval_last
            self.pval_crnt = self.init_pval_crnt
            self.p_series = [self.pval_last]
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            for i in range(1,len(self.w_series)):
                nu = self.w_series[i]-self.w_series[i-1]
                mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
                self.pval_last = self.pval_crnt
                alph = self.a*mu0*self.dt
                beta = self.b*nu

                self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
                self.p_series.append(self.pval_crnt)

        elif series == 'price':
            self.w_series = [0]

            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)

            pval_last = self.init_pval_last
            pval_crnt = self.init_pval_crnt

            for i in range(1,len(self.p_series)):

                pval_next = self.p_series[i]

                dP = (pval_next - pval_crnt)/(pval_crnt**(self.m))
                mu0 = (pval_crnt - pval_last)/pval_last
                self.w_series.append(self.w_series[-1]+(dP-self.a*mu0*self.dt)/self.b)

                pval_last = pval_crnt
                pval_crnt = pval_next
                    
    def analy_retrace(self, series, x=None):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            mu0 = 0
            count = 1
            for i in [self.dt*i for i in range(1,len(self.w_series))]:
                alph = (self.a*mu0-0.5*self.b**2)*i
                beta = self.b*self.w_series[count]
                self.p_series.append(self.init_pval_crnt*np.exp(alph + beta))

                mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
                count+=1
        
        if series == 'price':
            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)
            i = 1
            mu0 = 0
            self.w_series = [0]
            for t in [self.dt*j for j in range(1,len(self.p_series))]:
                alph = (self.a*mu0-0.5*self.b**2)*t
                self.w_series.append((np.log(float(self.p_series[i])/self.init_pval_crnt)-alph)/self.b)
                
                mu0 = (self.p_series[-1]-self.p_series[-2])/self.p_series[-2]
                i+=1
```

+ In Figure (2) it's quite clear that as time goes on the analytical `retrace()` of a price series produced using the `euler` method begins to diverge. This might be expected as they are different methods after all, however I believe that Figure (3) shows where this problem actually originates.
+ I think this is due to the fact that the solution given by equation 2.2 on page 40 in [A. J. Roberts' textbook](http://tinyurl.com/aj-roberts-txtbook) is not truly applicable to what I've done here. It comes down to the fact that the dominant term in the argument of the exponential function comes from $t$. In the book it discusses this and mentions that in the case where $\alpha$ is constant $X(t)$ eventually goes to zero. However in the particular case I have here, $\alpha = a \cdot \mu_0$ where $a$ is a constant and $\mu_0$ is equal to the instantaneous return rate, in other words $\mu_0 = \mu_0(\frac{dX}{dt})$. Thus as $t$ increases so does the influence of $\mu_0$ which is in turn strongly controled by variations in $W(t)$ resulting in the exponential growth of noise towards the end of the series.
+ In the end I think this simply means that I need to use a different analytical solution if $\alpha$ is ultimately dependant on $\frac{dX}{dt}$. Hopefully I can do that once I finnish chapter 2.
