```python
 def fast_group(a, t):
    # Anything greater than 0 should be 1.
    b = np.zeros(a.shape)
    b[a > 0.] = 1
    
    # Find leading and trailing edges.
    d = b - np.roll(b, -1)
    d = np.roll(d,1)
    
    # Get indicies of the edges.  Make sure 0 is an edge.
    e = np.where(d != 0)[0]
    if e[0] != 0:
        e = np.insert(e, 0, 0)
    
    # Calc the length of each segment usings edge indicies.
    f = np.roll(e, -1)
    f[-1] = a.shape[0]
    g = f - e
    
    # Repeat the edge indicies for the segment lengths.
    h = np.repeat(e, g)
    
    # Get the segment start times
    i = h - np.arange(a.shape[0])
    start = t[i + np.arange(a.shape[0])]
    
    # Return the delta from the last start time.
    return t - start
```

```python
def rolling_vwap(data,span):
    lst = []
    for i in range(span,len(data)):
        x = data[i-span:i]
        vwap = (x.close*x.volume).sum() / x.volume.sum()
        lst.append(vwap)
    
    dates = data[span:].index
    return pd.DataFrame(lst,columns = ['vwap'], index = dates)
```

```python
data = qqm.get_minutely_data('amzn','2013')[['open','high','low','close','volume']]
```

```python
vwap = rolling_vwap(data,10)
```

```python
sig = pd.rolling_std(data[['close']], 10)
sig.columns = ['sig']
```

```python
future_pct_change = pd.DataFrame((data.close.shift(-10)-data.close)/data.close)
future_pct_change = pd.DataFrame(future_pct_change)
future_pct_change.columns = ['future_pct_change']
```

```python
diff = vwap - data.close[len(data.close)-len(vwap):]
diff.columns = ['diff']
# abs_diff = abs(diff)
# abs_diff.columns = ['abs_diff']
# t = diff['diff'].index.values
# a = diff['diff'].values
# diff_duration = pd.DataFrame(fast_group(a,t).astype('int')/6e10,index = diff.index,columns = ['diff_duration'])
```

```python
delta = [10,20,30,60]
past_pct_change = pd.concat([np.log(data.close/data.close.shift(delta[0])),
                             np.log(data.close/data.close.shift(delta[1])),
                             np.log(data.close/data.close.shift(delta[2])),
                             np.log(data.close/data.close.shift(delta[3]))],axis = 1)
past_pct_change.columns = ['past_pct_change_60','past_pct_change_120','past_pct_change_240','past_pct_change_360']
```

```python
thresh = 0
thresh
```

```python
signal = pd.DataFrame(len(data)*'b'.split(),columns = ['signal'], index = future_pct_change.index)
signal[future_pct_change.values > thresh] = 'r' # Buy
signal[future_pct_change.values < -thresh] = 'g' # Sell
```

```python
new_data = pd.concat([data,diff,sig, future_pct_change,past_pct_change,signal],axis = 1)
new_data = new_data.dropna()
```
