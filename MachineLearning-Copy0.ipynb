```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
% matplotlib inline
%config InlineBackend.figure_format = 'svg'
import quantquoteminutely as qqm
import random
from sklearn.svm import LinearSVC
from sklearn import preprocessing
```

```python
def corrected_ewma(data,span):
    ewma = pd.stats.moments.ewma
    x = data
    
    # take EWMA in both directions with a smaller span term
    fwd = ewma( x, span=span ) # take EWMA in fwd direction
    bwd = ewma( x[::-1], span=span )[::-1] # take EWMA in bwd direction
    c = np.vstack(( fwd, bwd )) # lump fwd and bwd together
    c = np.mean( c, axis=0 ) # average

    return pd.DataFrame(c,columns = ['ewma'],index = data.index)
```

```python
 def fast_group(a, t):
    # Anything greater than 0 should be 1.
    b = np.zeros(a.shape)
    b[a > 0.] = 1
    
    # Find leading and trailing edges.
    d = b - np.roll(b, -1)
    d = np.roll(d,1)
    
    # Get indicies of the edges.  Make sure 0 is an edge.
    e = np.where(d != 0)[0]
    if e[0] != 0:
        e = np.insert(e, 0, 0)
    
    # Calc the length of each segment usings edge indicies.
    f = np.roll(e, -1)
    f[-1] = a.shape[0]
    g = f - e
    
    # Repeat the edge indicies for the segment lengths.
    h = np.repeat(e, g)
    
    # Get the segment start times
    i = h - np.arange(a.shape[0])
    start = t[i + np.arange(a.shape[0])]
    
    # Return the delta from the last start time.
    return t - start
```

```python
stock_list = np.array(qqm.get_symbols())
```

```python
sample = stock_list[random.sample(range(1, len(stock_list)), 100)]
```

```python
humongous_beeg_test = []

for symbol in sample:
    try:
        data = qqm.get_minutely_data(symbol)[['open','high','low','close','volume']]

        sig = pd.rolling_std(data[['close']], 120)
        sig.columns = ['sig']

        ewma = corrected_ewma(data.close,120)

        future_pct_change = pd.DataFrame((data.close.shift(-100)-data.close)/data.close)
        future_pct_change = pd.DataFrame(future_pct_change)
        future_pct_change.columns = ['future_pct_change']

        diff = ewma - data.close
        diff.columns = ['diff']

        past_pct_change = pd.concat([np.log(data.close/data.close.shift(60)),
                                     np.log(data.close/data.close.shift(120)),
                                     np.log(data.close/data.close.shift(240)),
                                     np.log(data.close/data.close.shift(360))],axis = 1)
        past_pct_change.columns = ['past_pct_change_60','past_pct_change_120','past_pct_change_240','past_pct_change_360']

        thresh = 0

        signal = pd.DataFrame(len(data)*'b'.split(),columns = ['signal'], index = future_pct_change.index)
        signal[future_pct_change.values > thresh] = 'r' # Buy
        signal[future_pct_change.values < -thresh] = 'g' # Sell

        new_data = pd.concat([data, ewma, diff, sig, future_pct_change, past_pct_change, signal],axis = 1)
        new_data = new_data.dropna()

        features = ['diff','sig','past_pct_change_60','past_pct_change_240']

        accuracy = []
        dates = []
        training_window = 10000
        test_window = 1000

        for i in range(1000):

                j = random.randint(training_window,len(new_data)-test_window-training_window)
                training_index = range(j-training_window,j)

                clf = LinearSVC(C = 0.01)
                test_index = range(j,j+test_window)
                x = new_data[features]
                dates.append(x.ix[training_index].index[0])

                X = preprocessing.scale(x.ix[training_index].values)
                y = new_data.signal.ix[training_index].values
                y = np.ravel(y, order='C')
                clf.fit(X,y)

                X_new = preprocessing.scale(new_data[features].ix[test_index].values)

                pred = pd.DataFrame(clf.predict(X_new),index = new_data.index[test_index])
                bools = pd.DataFrame(new_data.signal.ix[pred.index] == pred)
                ones = np.ones(len(bools))[np.ravel(bools.values)]

                accuracy.append(float(len(ones))/len(bools))

        accuracy = np.array(accuracy)

        humongous_beeg_test.append((accuracy.mean(),accuracy.std()))
    except:
        continue
```

```python
test = pd.DataFrame(humongous_beeg_test,columns = ['mu','sig'])
test.plot()
```
